```python
import numpy as np
import pandas as pd

from pandas import DataFrame
```

```python
numbers_df = DataFrame(np.arange(0,90,3).reshape(10,3), index=['row 1', 'row 2', 'row 3', 'row 4', 'row 5', 'row 6', 'row 7', 'row 8', 'row 9', 'row 10'], columns=['column 1', 'column 2', 'column 3'])
```

```python
numbers_df
```

```python
print(numbers_df.iloc[1,0])
```

```python
numbers_df.iloc[[6,7,9], [0,1,2]]
```

#### Comparison operators (> < = <= => == !=) and Masking.

```python
mask = numbers_df > 30

print(mask)
```

```python
mask
```

```python
numbers_df[numbers_df > 30] 
```
