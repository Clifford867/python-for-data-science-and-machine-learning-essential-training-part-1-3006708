## Working with Missing Data in Pandas

```python
import numpy as np
import pandas as pd

from pandas import DataFrame
```

```python
data = {'Names': ["Steve", "John", "Richard", "Sarah", "Randy", "Micheal", "Julie"],
        "Age": [20, 22, 20, 21, 24, 23, 22], 
        "Gender": ["Male", "Male", "Male", "Female", "Male", "Male", "Female"],
        "Rank": [2, 1, 4, 5, 3, 7, 6]}
```

```python
ranking_df = DataFrame(data)
```

```python
ranking_df
```

```python
ranking_df.iloc[2:5,1] = np.nan
ranking_df.iloc[3:6,3] = np.nan
ranking_df.iloc[3,:] = np.nan
ranking_df
```

```python
ranking_df.isnull()
```

```python
ranking_df.notnull()
```

```python
rank_mask = ranking_df["Rank"].isnull()

rank_mask
```

```python
ranking_df[rank_mask]
```

```python
bool_series = pd.isnull(ranking_df["Age"])

bool_series
```

```python
ranking_df[bool_series]
```

```python
ranking_df.fillna(0)
```

```python
ranking_df
```

```python
ranking_df.ffill()
```

```python
ranking_df.bfill()
```

```python
ranking_df.dropna(how='all')
```

```python
cliff_df = ranking_df.copy()
cliff_df
```

```python
cliff_df.iloc[:,2] = np.nan
cliff_df
```

```python
cliff_df.dropna(how='all')
```

```python
cliff_df.dropna(axis=1)
```

```python
ranking_df.dropna(axis=0)
```
